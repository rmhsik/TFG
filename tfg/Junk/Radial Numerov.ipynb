```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
me = 1
mp = 1836.15
hbar = 1
mu = (me*mp)/(me+mp)
Z = 1
```

```python
fig = plt.figure();
ax = fig.add_subplot(111);

r0 = 1E-3
rf = 7
N = 1000

r = np.linspace(r0,rf,2*N)
PsiR = np.zeros(N)
PsiL = np.zeros(N)

l = 0
Veff = np.zeros(2*N)
for i in range(len(r)):
    Veff[i] = -Z/r[i]+(l*(l+1))/(2*mu*r[i]**2)

ep = -
k2 = 2*mu/hbar**2*(ep-Veff)
h2 = (abs(r0-rf)/(2*N))**2

def WaveFunction(k2,N,psil,psir): ### FINALLY IT WORKS!!
    psil[0]=0
    psil[1] = 1E-7
    psir[-1] = 0
    psir[-2] = 1E-7 # negative for odd parity eigenfunctions
    for i in range(2,N):
        psil[i] = (2*(1-(5.0/12)*h2*k2[i-1])*psil[i-1]-(1+(1.0/12)*h2*k2[i-2])*psil[i-2])/(1+(1.0/12)*h2*k2[i])
    for i in range(3,N): #The minus in indexes indicates that we start from right to the left
        psir[-i] = (2*(1-(5.0/12)*h2*k2[-i+1])*psir[-i+1]-(1+(1.0/12)*h2*k2[-i+2])*psir[-i+2])/(1+(1.0/12)*h2*k2[-i])
    return psil,psir

psi0 = WaveFunction(k2,N,PsiL,PsiR)
ax.plot(r[0:N],psi0[0]/r[0:N])
ax.plot(r[N-1:-1],psi0[1]/r[N-1:-1])
#ax.plot(r,Veff)
#ax.plot(r,ep*np.ones(2*N),'--')
#ax.plot(x,-0.85*np.ones(2*N),'--')
plt.show()

ax.plot(r,k2)
plt.show()
```
