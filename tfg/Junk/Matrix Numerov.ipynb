```python
import Math as mt
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

```python
N = 400
xmin = -2.5
xmax = 2.5

x = np.linspace(-2.5,2.5,400)
h = (xmax-xmin)/N
v = 0.5*x**2

A = (1/h**2)*(mt.sdiag(N,-1)-2*mt.diag(N)+mt.sdiag(N,1))
B = (1/12)*(mt.sdiag(N,-1)+10*mt.diag(N)+1*mt.sdiag(N,1))
V = v*mt.diag(N)
```

```python
Binv = linalg.inv(B)
H = -1/2*np.dot(Binv,A) + V
```

```python
fig = plt.figure();
ax = fig.add_subplot(111);

ax.plot(x,v)
```

```python
fig = plt.figure();
ax = fig.add_subplot(111);

ax.plot(x,linalg.eig(H)[1][-1])

```
